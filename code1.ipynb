```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings(action="ignore")
plt.style.use(["seaborn-bright","dark_background"])
```

```python
data = pd.read_csv("/content/COVID-19 Survey Student Responses.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data.dropna(axis = 0,inplace=True)
```

From which region the most of the students are

```python
fig = px.pie(data, names="Region of residence", color="Region of residence",height=500,width=700)
fig.show()
```

Which age group they belong to?

```python
fig = px.histogram(data, x = "Age of Subject",nbins=25,marginal="box")
fig.show()
```

Most of the students belong to age group 15-24.

How much time they spent on online class?

```python
fig = px.bar(data, x = "Time spent on Online Class",y = "Age of Subject")
fig.show()
```

```python
fig = px.pie(data, names="Rating of Online Class experience",color = "Rating of Online Class experience")
fig.show()
```

```python
data["Time spent on TV"].unique()
```

Are they connected with family , friends and relatives?

```python
data["Are they connected with family and friends"] = data["Do you find yourself more connected with your family, close friends , relatives  ?"]
data.drop(columns = ["Do you find yourself more connected with your family, close friends , relatives  ?"],inplace= True)
```

```python
df = data.groupby(by=["Health issue during lockdown", "Are they connected with family and friends", "Change in your weight","Number of meals per day"]).count()[["ID"]].rename(columns={"ID":"Count"})
df["Total"] = "Total"
df = df.reset_index()
df.head()
```

```python
fig = px.sunburst(df,
                  path=["Total","Health issue during lockdown", "Are they connected with family and friends", "Change in your weight","Number of meals per day"],
                  values='Count',
                  title = "Sunbrust Path [Total,Health issue during lockdown,Are they connected with family and friends,Change in your weight,Number of meals per day]",
)
fig.show()
```
